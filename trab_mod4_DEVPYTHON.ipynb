```
import threading
import time
from random import randint
```

```
def funcao_1(num):
    n=num
    while n>0:
      n-=1
      print("n_1: {}".format(n))
      time.sleep(randint(0,2))
      
```

```
def funcao_2(num):
    n=num

    while n<100:
      n+=1
      print("n_2: {}".format(n))
      time.sleep(randint(0,2))
```

```
if  __name__ == "__main__":
  t1 = threading.Thread(target=funcao_1, args=(100,))
  t2 = threading.Thread(target=funcao_2, args=(0,))

  t1.start()
  t2.start()

  t1.join()
  t2.join()
```
