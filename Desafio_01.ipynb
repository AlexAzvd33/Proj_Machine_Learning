```
class Classe_1:
    def funcao_da_classe_1(self, int):
        dicionario = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C':100, 'D': 500, 'M': 1000}
        valor = 0
        for i in range(len(int)):
            if i > 0 and dicionario[int[i]] > dicionario[int[i - 1 ]]:
                valor += dicionario[int[i]] - 2 * dicionario[int[i - 1]]
            else:
                valor += dicionario[int[i]]
        return valor
```
